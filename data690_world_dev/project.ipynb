### I have taken 2019 data

```python
df_2019 = df.query("Year == 2019")
```

```python
df_2019.sample(10)
```

```python
df_2019.shape
```

### Finding GDP of the countries using 50% of median people in 2019

```python
df_2019_gdp = df_2019.query("indicator in['NY.GDP.PCAP.KD.ZG']")
df_2019_gdp
```

### Bar graph of Countries by 2019 GDP per Capita

```python
fig = px.bar(
    data_frame=df_2019_gdp,
    title="Countries by 2019 GDP per Capita",
    labels={"value":"GDP per Capita"},
    x="Country Name",
    y="value",
    color="Country Name"
)

fig.update_xaxes(categoryorder="total descending")


fig.show()
```

### Getting the data of 50% median people in the country using the poverty indicator

```python
df_2019_me = df_2019.query("indicator in['SI.DST.50MD']")
df_2019_me
```

### Bar graph Countries by 2019 50% median income people

```python
fig = px.bar(
    data_frame=df_2019_me,
    title="Countries by 2019 50% median income people",
    labels={"value":"GDP per Capita", "Year":"YEAR"},
    x="Country Name",
    y="value",
    color="Country Code"
)

fig.update_xaxes(categoryorder="total descending")


fig.show()
```

### Considering year 2010 data from the data we have 

```python
df_2010 = df.query("Year == 2010")
df_2010.sample(10)
```

```python
df_2010.shape
```

```python
df_2010_me = df_2010.query("indicator in['SI.DST.50MD']")
df_2010_me
```

### Pie chart of Countries by 2010 of 50% median income (% of GDP)

```python
fig = px.pie(
    df_2010_me,
    values='value', 
    names='Country Name',
    title='Countries by 2010 of 50% median income (% of GDP)',
    hover_name='Country Code',
)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
df_2015_me = df_2015.query("indicator in['SI.DST.50MD']")
df_2015_me
```

### Line plot for Countries by 50% of median income (% of GDP)

```python
fig = px.line(
    df_me,
    x="Year",
    y="value",
    title="Countries by 50% of median income (% of GDP)",
    labels={"value":"Total 50% of median income people"},
    color='Country Name' 
)

fig.update_layout(showlegend=False)

fig.show()
```
